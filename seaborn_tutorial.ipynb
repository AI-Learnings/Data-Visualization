# Controlling Ploted Figure Aesthetics
1. Figure styling
2. Axes styling
3. Color Palettes etc

```
def sinplot(flip=1):
  x = np.linspace(0,14,200)
  for i in range(1,7):
    plt.plot(x, np.sin(x+i*0.5)*(7-i)*flip)
```

```
sinplot()
```

```
sns.set_style('white')
sinplot()
sns.despine(left=True)
```
